```python
mainPageDoc = pq(res.text)
#爬取頁面所有標題和作者
# 做迭代
# 從標題移動到作者
for each in mainPageDoc("#main-container  div > div.title > a").items():
    print(each.text(), each.parent().siblings(".meta>.author").text())     #加上標題
    
#跳到title div這個標籤 ，找出兄弟姐妹meta :siblings(選擇器)  ，再找meta底下的author這個類別(>.author)  
```

```python
# 讓爬蟲可以翻動頁面(不只爬第一頁) => 把上一頁這個標籤的程式複製起來:(確認每頁的按鈕程式都一樣)
# #action-bar-container > div > div.btn-group.btn-group-paging > a:nth-child(2)
```

```python
for each in mainPageDoc("#main-container  div > div.title > a").items():
    print(each.text(), each.parent().siblings(".meta>.author").text())     #加上標題
    
    
print(  mainPageDoc("#action-bar-container > div > div.btn-group.btn-group-paging > a:nth-child(2)") .attr("href") ) #裝"上一頁"這個標籤的程式複製起來
# a標籤想要拿到"href"屬性裡面的值的話: .attr("href")")
# 把當下這頁的上一頁連結印出來看(相對路徑)
```

```python
for each in mainPageDoc("#main-container  div > div.title > a").items():
    print(each.text(), each.parent().siblings(".meta>.author").text())     #加上標題
    
mainPageDoc.make_links_absolute(base_url=res.url) #取得網頁是由requests進行，不是pq: (base_url=res.url)
print(  mainPageDoc("#action-bar-container > div > div.btn-group.btn-group-paging > a:nth-child(2)") .attr("href") ) #"下一頁按鈕"的程式

#最後一行的連結變成絕對路徑了(抓取上一頁按鈕的連結)
```
